```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel('edutech companies.xlsx')
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
x = data['Category'].value_counts()
```

## Category based analysis

### Category wise distribution of given data

```python
plt.figure(figsize=(20,10))
colors = sns.color_palette('bright')
lables = ['Data Science','Software Eng','School edu','IIT JEE/NEET','GATE','Kids edu','Mentorship and Counselling']
plt.pie(x,labels = lables, colors = colors, autopct = '%0.0f%%')
plt.show()
```

### which category has over all more revenue generation
we can say IIT JEE/NEET classes are generating more income according given data

```python
data.groupby('Category')['Turnover (in Cr)'].sum().sort_index().plot(kind='pie',figsize=(20,10),title = "which category has over all more revenue generation")
```

### Rating based on catagories
ratings start to fall when it comes to school education and software engg

```python
data.groupby('Category')['Rating'].max().plot(kind='line',figsize=(20,8),title = "Ratings based on categories",marker='D')
```

### No of companies under a category

```python
x.sort_index().plot(kind='bar',figsize=(20,8),title = "most top ratings",color='green')
```

### in which year most of the given companies where established
since 2014 we can say the evolution of establishment took place

```python
data['Year of Establishment'].value_counts().nlargest(10).plot(kind='pie',figsize=(20,10),title = "top years of establishment with most establishment")
```

### turn over earned by company according establishment year
we can say that turn over earned is directly related to establishment year

```python
data.groupby('Company Name')['Turnover (in Cr)'].sum().sort_values(ascending=False).head(10).plot(kind='pie',figsize=(20,10),title = "most turnovers in cr made by most companies")
```

### in which year which category is Established more

```python
data.groupby('Year of Establishment')['Category'].value_counts().nlargest(10).sort_index().plot(kind='pie',figsize=(24,10),title = "most top ratings")
```

### fluctuation in Category according years

```python
data.groupby('Year of Establishment')['Category'].max().value_counts().sort_index().plot(kind='line',figsize=(20,8),title = "most top ratings",color='orange')
```

```python
data.groupby('Year of Establishment')['Category'].value_counts().nlargest(20).sort_index().plot(kind='bar',figsize=(20,8),title = "most top ratings",color='orange')
```
